## NIKE Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load your stock market data into a pandas DataFrame
data = pd.read_csv("NKE.csv", index_col="Date", parse_dates=True)

# **1. Descriptive Statistics:**

# Calculate mean, median, and standard deviation
data.describe()
```

```python
# **Plots:**

# Boxplot
data.plot(kind="box")
plt.title("Boxplot of Stock Prices")
plt.show()

# Histogram
data["Close"].hist()
plt.title("Histogram of Closing Price")

plt.show()

```

```python
# **2. Time Series Analysis:**

# Daily percentage change
data["Change"] = ((data["Close"] - data["Close"].shift(1)) / data["Close"].shift(1)) * 100

# Moving averages
data["SMA50"] = data["Close"].rolling(window=50).mean()
data["EMA200"] = data["Close"].ewm(alpha=0.5, min_periods=200, adjust=False).mean()
data


```

```python
# **Plots:**

# Line chart
plt.plot(data["Close"], label="Close")
plt.plot(data["SMA50"], label="SMA50")
plt.plot(data["EMA200"], label="EMA200")
plt.title("Closing Price, SMA, and EMA")
plt.legend()
plt.show()

# Volume plot
plt.plot(data["Volume"])
plt.title("Daily Volume")
plt.show()

```

```python
# **3. Volatility:**

# Standard deviation
data["Volatility"] = data["Change"].std() * np.sqrt(252)  # Annualize daily volatility
data

```

```python
# Average True Range (ATR)
def calculate_atr(high, low, close, window=14):
    tr = pd.DataFrame()
    tr["HL"] = high - low
    tr["HC"] = abs(high - close.shift())
    tr["LC"] = abs(low - close.shift())
    tr["TR"] = tr.max(axis=1)
    atr = tr["TR"].rolling(window=window).mean()
    return atr

data["ATR"] = calculate_atr(data["High"], data["Low"], data["Close"])
data


```

```python
# **Plots:**

# Line chart
plt.plot(data["Volatility"], label="Volatility")
plt.plot(data["ATR"], label="ATR")
plt.title("Volatility Measures (Annualized SD & ATR)")
plt.legend()
plt.show()
```

```python


# Assuming you have your data loaded in the 'data' DataFrame
# Replace 'your_ticker_symbol' with the actual symbol
your_ticker_symbol = 'AAPL'  # Example ticker symbol

# Get the first date from your data
first_date = data.index[0]

# Filter data from the first date
filtered_data = data.loc[first_date:]

# Plot the candlestick chart
mpf.plot(filtered_data, type='candle', title="Candlestick Chart", ylabel='Price', style='charles')
```

```python
df_nike['Date'] = pd.to_datetime(df_nike['Date'])

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(df_nike['Date'], df_nike['Open'], label='Open')
plt.plot(df_nike['Date'], df_nike['High'], label='High')
plt.plot(df_nike['Date'], df_nike['Low'], label='Low')
plt.plot(df_nike['Date'], df_nike['Close'], label='Close')
plt.plot(df_nike['Date'], df_nike['Adj Close'], label='Adj Close')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Nike Stock Data Over Time')
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()

# Plotting Volume separately
plt.figure(figsize=(10, 3))

plt.plot(df_nike['Date'], df_nike['Volume'], label='Volume', color='black')

plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Volume Over Time')
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
## Now creating the regression model for tclosing prices 
df_nike.drop("Date",axis=1,inplace=True)
X=df_nike.drop("Close",axis=1)
Y=df_nike['Close']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
X_train.shape,Y_train.shape




lr=LinearRegression()
lr.fit(X_train,Y_train)

accuracy=lr.score(X_test,Y_test)
print("Accuracy on test data ",accuracy)
```

### COMPARISON of Closing Price and SMA

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data into pandas DataFrames
data1 = pd.read_csv("NKE.csv", index_col="Date", parse_dates=True)
data2 = pd.read_csv("ADS.DE.csv", index_col="Date", parse_dates=True)

# **1. Candlestick Chart Comparison:**

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(15, 6))

# Plot closing prices and candlesticks for Dataset 1 on ax1
ax1.plot(data1["Close"], label="Dataset 1 Close")

# Create candlestick bars using efficient vectorized operations
close_diff = data1["Close"] - data1["Open"]
colors = np.where(close_diff > 0, "green", "red")
ax1.bar(data1.index, close_diff, color=colors, width=0.6, bottom=data1["Low"])
ax1.bar(data1.index, data1["Open"], color=colors, width=0.6, bottom=data1["Low"])
ax1.legend()

# Similar approach for Dataset 2 on ax2
ax2.plot(data2["Close"], label="Dataset 2 Close")
close_diff = data2["Close"] - data2["Open"]
colors = np.where(close_diff > 0, "green", "red")
ax2.bar(data2.index, close_diff, color=colors, width=0.6, bottom=data2["Low"])
ax2.bar(data2.index, data2["Open"], color=colors, width=0.6, bottom=data2["Low"])
ax2.legend()

# Adjust layout and styling
plt.suptitle("Comparison of Closing Prices and SMA")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
